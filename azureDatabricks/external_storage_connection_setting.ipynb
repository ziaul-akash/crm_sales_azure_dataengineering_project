```python
# Set the access key for the storage account
#we need to create scope in databricks and connect it with azure key_vault secret to access data from storage

storage_account_key = dbutils.secrets.get(
    scope="testScope",
    key="test-keyvault"
)
spark.conf.set(
    "fs.azure.account.key.ziaulhaqakash.dfs.core.windows.net",
    storage_account_key
)

test_df = spark.read.format("csv").option("header", "true").option("inferSchema", "true").load(
    "abfss://test@ziaulhaqakash.dfs.core.windows.net/sales_teams.csv/"
)

test_df.write.format("csv").mode('overwrite').option('header', 'true').save(
    "abfss://transform@ziaulhaqakash.dfs.core.windows.net/sales_teams/"
)
```
